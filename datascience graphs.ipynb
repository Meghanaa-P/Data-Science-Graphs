```python
import matplotlib.pyplot as plt
import numpy as np

x = ['python','c','c++','java']
y = [85,70,60,82]
z = [20,30,40,50]

width=0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("no.",fontsize=15)
plt.title("DS",fontsize=15)

plt.bar(p,y,width,color='y',label='datascience')
plt.bar(p1,z,width,color='r',label='datascience1')

plt.legend()
plt.show()
```
